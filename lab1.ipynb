##### 1.1. Cars characteristics

```python
# models
most_expensive_car = df['model'][df['price'] == max(df['price'])].values[0]
least_expensive_car = df['model'][df['price'] == min(df['price'])].values[0]
most_recent_car = df['model'][df['year'] == max(df['year'])].values[0]
oldest_car = df['model'][df['year'] == min(df['year'])].values[0]
most_mileage = df['model'][df['mileage'] == max(df['mileage'])].values[0]
least_mileage = df['model'][df['mileage'] == min(df['mileage'])].values[0]

# numbers
most_expensive_price = max(df['price'])
least_expensive_price = min(df['price'])
most_recent_car_year = max(df['year'])
oldest_car_year = min(df['year'])
most_mileage_km = max(df['mileage'])
least_mileage_km = min(df['mileage'])

# Combined
print(f"Most expensive SUV: {most_expensive_car} - {most_expensive_price}")
print(f"Least expensive SUV: {least_expensive_car} - {least_expensive_price}")
print(f"Mean price: {int(df['price'].mean())} euros")
print(f"Most recent SUV: {most_recent_car} - {most_recent_car_year}")
print(f"Oldest SUV: {oldest_car} - {oldest_car_year}")
print(f"SUV with most mileage: {most_mileage} - {most_mileage_km}")
print(f"SUV with least mileage: {least_mileage} - {least_mileage_km}")
```

##### 2. Creating plots

```python
plt.figure(figsize=(10,6))
plt.scatter(df['mileage'],df['price'])
plt.xlabel('Mileage(km)', fontsize = '12')
plt.ylabel('Price(eur)', fontsize = '12')
plt.title("Car mileage vs price", fontsize = 15)
plt.grid(True)
plt.show()
```

##### 3. Checking how many prices are above the threshold

```python
threshold = int(df['price'].mean()) 

above_threshold = len(df[df['price'] > threshold])
cars_above_threshold = df['model'][df['price'] > threshold].values

print(f"{above_threshold} cars are above the threshold price:\n" + "\n".join(cars_above_threshold))
```

##### 4. Computing the mean and standard deviation of the cost. Printing the results as boxplot

```python
mean_price = int(df['price'].mean())
median_price = int(df['price'].median())
standard_deviation = round(df['price'].std(),2)
print(f"Mean Price: {mean_price}")
print(f"Median Price: {median_price}")
print(f"Standard Deviation: {standard_deviation}")
```

```python
# creating a boxplot of the car prices
plt.figure(figsize=(10, 7))
sns.boxplot(data=df['price'], color='lightblue')

# adding lines for mean and standard deviation (axhline = add horizontal line across the Axes)
plt.axhline(mean_price, color='red', linestyle='--', label=f'Mean: {mean_price}')
plt.axhline(mean_price + standard_deviation, color='green', linestyle='--', label=f'Mean + Std: {mean_price + standard_deviation}')
plt.axhline(mean_price - standard_deviation, color='green', linestyle='--', label=f'Mean - Std: {mean_price - standard_deviation}')

# adding title and labels
plt.title('Price Distribution of SUVs', fontsize=14)
plt.ylabel('Price (in EUR)', fontsize=12)

plt.legend()
plt.show()
```
